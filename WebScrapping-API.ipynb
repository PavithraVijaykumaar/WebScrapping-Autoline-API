# Web Scrapping from Autoline website

Autoline website which contains the details of used `cars` around the mentioned city is used for scrapping. 

This method of scrapping uses `cURL` for sending HTTP requests to retrieve the data from the website.

## Importing Libraries

```python
import requests
import pandas as pd
```

## Requesting connection for the first city

```python
headers = {
    'Accept': '*/*',
    'Accept-Language': 'en-GB,en;q=0.9,ta-IN;q=0.8,ta;q=0.7,en-US;q=0.6',
    'Connection': 'keep-alive',
    'If-None-Match': 'W/^\\^e3211357bab550aa0622b8cbd5cf050e^\\^',
    'Origin': 'https://www.autolist.com',
    'Referer': 'https://www.autolist.com/',
    'Sec-Fetch-Dest': 'empty',
    'Sec-Fetch-Mode': 'cors',
    'Sec-Fetch-Site': 'same-site',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36',
    'sec-ch-ua': '^\\^Not.A/Brand^\\^;v=^\\^8^\\^, ^\\^Chromium^\\^;v=^\\^114^\\^, ^\\^Google',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '^\\^Windows^\\^',
}

params = (
    ('city', 'California^%^20City'),
    ('latitude', '35.125801'),
    ('limit', '20'),
    ('longitude', '-117.9859038'),
    ('make', 'Ford'),
    ('page', '1'),
    ('radius', '100'),
    ('state', 'CA'),
)

response = requests.get('https://production-proxy.autolist.com/api/cwv/seo/listings', headers=headers, params=params)
```

## Multiple page Web Scrapping

```python
#scrapping multiple webpages
headers = {
    'Accept': '*/*',
    'Accept-Language': 'en-GB,en;q=0.9,ta-IN;q=0.8,ta;q=0.7,en-US;q=0.6',
    'Connection': 'keep-alive',
    'If-None-Match': 'W/^\\^e3211357bab550aa0622b8cbd5cf050e^\\^',
    'Origin': 'https://www.autolist.com',
    'Referer': 'https://www.autolist.com/',
    'Sec-Fetch-Dest': 'empty',
    'Sec-Fetch-Mode': 'cors',
    'Sec-Fetch-Site': 'same-site',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36',
    'sec-ch-ua': '^\\^Not.A/Brand^\\^;v=^\\^8^\\^, ^\\^Chromium^\\^;v=^\\^114^\\^, ^\\^Google',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '^\\^Windows^\\^',
}

model = []
year = []
dealer = []
mileage = []
price = []
city=[]

for i in range(1,634):
    
    params = (
        ('city', 'California^%^20City'),
        ('latitude', '35.125801'),
        ('limit', '20'),
        ('longitude', '-117.9859038'),
        ('make', 'Ford'),
        ('page', str(i)),
        ('radius', '100'),
        ('state', 'CA'),
    )
    #variable for obtaining request
    response = requests.get('https://production-proxy.autolist.com/api/cwv/seo/listings', headers=headers, params=params)
    
    #json object
    data_json=response.json()
    
    #getting 20 items for each page
    data_items=data_json['search_results']
    
    for data in data_items:
        model.append(data['model'])
        year.append(data['year'])
        dealer.append(data['dealer_name'])
        mileage.append(data['mileage'])
        price.append(data['price'])
        city.append(data['city'])
    
```

After scrapping off entire pages, the retrieved datas are stored in a `DataFrame` format named as `ford_pages_df`.

## Converting into DataFrame

```python
ford_pages_df=pd.DataFrame({'Model': model,'Mileage': mileage,'City': city,'Year': year,'Dealer': dealer,'Price': price})
```

## Multiple page Web Scrapping

```python
#multiple page scrapping for second city
headers = {
    'sec-ch-ua': '^\\^Not.A/Brand^\\^;v=^\\^8^\\^, ^\\^Chromium^\\^;v=^\\^114^\\^, ^\\^Google',
    'Referer': 'https://www.autolist.com/',
    'sec-ch-ua-mobile': '?0',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36',
    'sec-ch-ua-platform': '^\\^Windows^\\^',
}

model = []
year = []
dealer = []
mileage = []
price = []
city=[]

for i in range(1,188):
    params = (
        ('include_total_price_change', 'true'),
        ('include_time_on_market', 'true'),
        ('include_relative_price_difference', 'true'),
        ('latitude', '42.8286091'),
        ('limit', '20'),
        ('longitude', '-98.3298021'),
        ('make', 'Ford'),
        ('page', '1'),
        ('radius', '100'),
        ('zip', '600099'),
    )

    response = requests.get('https://production-proxy.autolist.com/api/v2/search', headers=headers, params=params)

    #json object
    data_json=response.json()
    
    #getting 20 items for 187 page
    data_items=data_json['records']
    
    for data in data_items:
        model.append(data['model_name'])
        year.append(data['year'])
        dealer.append(data['dealer_name'])
        mileage.append(data['mileage'])
        price.append(data['price'])
        city.append(data['city'])
```

After scrapping off entire pages, the retrieved datas are stored in a `DataFrame` format named as `ford_pages2_df`.

## Converting into DataFrame

```python
ford_pages2_df=pd.DataFrame({'Model': model,'Mileage': mileage,'City': city,'Year': year,'Dealer': dealer,'Price': price})
```

```python
ford_pages2_df
```

The entire scrapped datas are a count of `3740` cars from `187` pages

Now the obtained dataframes for are both the cities are combined using `concat` function and stored as an `excel` sheet

## Merging of DataFrame 

```python
#merging both dataframes
merged=pd.concat([ford_pages_df,ford_pages2_df],ignore_index=True)
```
