# Retrieving data from LocalHost API

The scrapped data which is stored in excel file is converted into `csv` format and a `LocalHost API` is created using `FastAPI` 

## Importing necessary libraries

```python
import pandas as pd
import requests
```

Retrieving data from the created LocalHost using `requests` library

```python
response= requests.get('http://127.0.0.1:8000/data')
```

```python
response
```

This `<Response [200]>` means that a successfull connection is made to the LocalHost for retrieving the data

```python
data=response.text
data
```

```python
data=response.json()
```

```python
df=pd.read_json(response.text)
df.head(5)
```
